```python
import matplotlib.pyplot as plt
import pickle
import numpy as np

with open('loss/nmt_posEmb.pkl', 'rb') as f:
    data_emb = pickle.load(f)

with open('loss/nmt_posEnc.pkl', 'rb') as f:
    data_enc = pickle.load(f)

x = list(range(1, 55))
l = len(x)
train_loss = [data_emb['train_loss_history'][:l], data_enc['train_loss_history'][:l]]
val_loss = [data_emb['val_loss_history'][:l], data_enc['val_loss_history'][:l]]
b2 = [data_emb['val_score_history']['bleu2'][:l], data_enc['val_score_history']['bleu2'][:l]]
b4 = [data_emb['val_score_history']['bleu4'][:l], data_enc['val_score_history']['bleu4'][:l]]
n2 = [data_emb['val_score_history']['nist2'][:l], data_enc['val_score_history']['nist2'][:l]]
n4 = [data_emb['val_score_history']['nist4'][:l], data_enc['val_score_history']['nist4'][:l]]
print(data_emb.keys())

plt.rcParams['ps.useafm'] = True
plt.rcParams['pdf.use14corefonts'] = True
plt.rcParams['text.usetex'] = True
plt.figure(figsize=(12, 8))
plt.rc('xtick', labelsize=15)
plt.rc('ytick', labelsize=15)
plt.rc('legend', fontsize=12)
plt.plot(x, train_loss[0], '-o', label='Positional Embedding')
plt.plot(x, train_loss[1], '-o', label='Positional Encoding')
plt.title('Training Set Loss', fontsize=20)
plt.xlabel('Epochs', fontsize=20)
plt.ylabel('Loss', fontsize=20)
plt.legend()
plt.grid()
plt.show()


plt.rcParams['ps.useafm'] = True
plt.rcParams['pdf.use14corefonts'] = True
plt.rcParams['text.usetex'] = True
plt.figure(figsize=(12, 8))
plt.rc('xtick', labelsize=15)
plt.rc('ytick', labelsize=15)
plt.rc('legend', fontsize=12)
plt.plot(x, val_loss[0], '-o', label='Positional Embedding')
plt.plot(x, val_loss[1], '-o', label='Positional Encoding')
plt.title('Test Set Loss', fontsize=20)
plt.xlabel('Epochs', fontsize=20)
plt.ylabel('Loss', fontsize=20)
plt.legend()
plt.grid()
plt.show()

plt.rcParams['ps.useafm'] = True
plt.rcParams['pdf.use14corefonts'] = True
plt.rcParams['text.usetex'] = True
plt.figure(figsize=(12, 8))
plt.rc('xtick', labelsize=15)
plt.rc('ytick', labelsize=15)
plt.rc('legend', fontsize=12)
plt.plot(x, np.exp(val_loss[0]), '-o', label='Positional Embedding')
plt.plot(x, np.exp(val_loss[1]), '-o', label='Positional Encoding')
plt.title('Test Set Score', fontsize=20)
plt.xlabel('Epochs', fontsize=20)
plt.ylabel('Perplexity (PPL)', fontsize=20)
plt.legend()
plt.grid()
plt.show()


plt.rcParams['ps.useafm'] = True
plt.rcParams['pdf.use14corefonts'] = True
plt.rcParams['text.usetex'] = True
plt.figure(figsize=(12, 8))
plt.rc('xtick', labelsize=15)
plt.rc('ytick', labelsize=15)
plt.rc('legend', fontsize=12)
plt.plot(x, b2[0], '-o', label='Positional Embedding')
plt.plot(x, b2[1], '-o', label='Positional Encoding')
plt.title('Test Set Score', fontsize=20)
plt.xlabel('Epochs', fontsize=20)
plt.ylabel('BLEU-2', fontsize=20)
plt.legend()
plt.grid()
plt.show()

plt.rcParams['ps.useafm'] = True
plt.rcParams['pdf.use14corefonts'] = True
plt.rcParams['text.usetex'] = True
plt.figure(figsize=(12, 8))
plt.rc('xtick', labelsize=15)
plt.rc('ytick', labelsize=15)
plt.rc('legend', fontsize=12)
plt.plot(x, b4[0], '-o', label='Positional Embedding')
plt.plot(x, b4[1], '-o', label='Positional Encoding')
plt.title('Test Set Score', fontsize=20)
plt.xlabel('Epochs', fontsize=20)
plt.ylabel('BLEU-4', fontsize=20)
plt.legend()
plt.grid()
plt.show()



plt.rcParams['ps.useafm'] = True
plt.rcParams['pdf.use14corefonts'] = True
plt.rcParams['text.usetex'] = True
plt.figure(figsize=(12, 8))
plt.rc('xtick', labelsize=15)
plt.rc('ytick', labelsize=15)
plt.rc('legend', fontsize=12)
plt.plot(x, n2[0], '-o', label='Positional Embedding')
plt.plot(x, n2[1], '-o', label='Positional Encoding')
plt.title('Test Set Score', fontsize=20)
plt.xlabel('Epochs', fontsize=20)
plt.ylabel('NIST-2', fontsize=20)
plt.legend()
plt.grid()
plt.show()



plt.rcParams['ps.useafm'] = True
plt.rcParams['pdf.use14corefonts'] = True
plt.rcParams['text.usetex'] = True
plt.figure(figsize=(12, 8))
plt.rc('xtick', labelsize=15)
plt.rc('ytick', labelsize=15)
plt.rc('legend', fontsize=12)
plt.plot(x, n4[0], '-o', label='Positional Embedding')
plt.plot(x, n4[1], '-o', label='Positional Encoding')
plt.title('Test Set Score', fontsize=20)
plt.xlabel('Epochs', fontsize=20)
plt.ylabel('NIST-4', fontsize=20)
plt.legend()
plt.grid()
plt.show()


```
